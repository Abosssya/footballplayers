```python
# Importing libraries and data
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
# reading data
df = pd.read_csv('/Users/abay/Downloads/fifa_players.csv')
```

The most expensive player

```python
df_expensive_player = df.groupby('name')[['value_euro','overall_rating']].agg('max')
#sorting to find the most expensive
df_expensive_player = df_expensive_player.sort_values(by='value_euro',ascending=False).head(1)

df_expensive_player
```

# Analyze the relationship between transfer fees and player performance.

```python
df_expensive_player_10 = df.groupby('name')[['value_euro','overall_rating']].agg('max')
#sorting values
df_expensive_player_10 = df_expensive_player_10.sort_values(by='value_euro',ascending=False).head(10)

# Plotting
df_expensive_player_10.plot(kind='barh',legend=False)
plt.title('Top 10 players fee')
plt.xlabel('Millions Euro')
plt.ylabel('Player')


# 1e8 = 1 * 10^8 = 100 000 000
```
