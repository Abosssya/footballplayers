```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/abay/Downloads/fifa_players.csv')
df.columns
```

```python
#The most expensive player
```

```python
df_expensive_player = df.groupby('name')[['value_euro','overall_rating']].agg('max').sort_values(by='value_euro',ascending=False).head(1)
df_expensive_player
```

# Analyze the relationship between transfer fees and player performance.

```python
df_expensive_player_10 = df.groupby('name')[['value_euro','overall_rating']].agg('max').sort_values(by='value_euro',ascending=False).head(10)
df_expensive_player_10.plot(kind='barh')
plt.title('Top 10 players fee')
```
